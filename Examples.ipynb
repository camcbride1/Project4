```python
# Setup

import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import random as rn
import seaborn as sns
from mpl_toolkits.basemap import Basemap
%matplotlib inline
plt.style.use('dark_background')
```

# Simple Line Plots

```python
# Simple Line Plots-1

# Setup
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0, 5, 1000)
# Plot
plt.plot(x, np.tan(x))
plt.plot(x, 1/(np.tan(x) + 1))
plt.show()
```

```python
# Simple Line Plots-2

# Plotting a complete simple line plot.

# Plot four lines
plt.plot(x, x, linestyle='solid', label= 'x')
plt.plot(x, 2 * x, linestyle='dashed', label= '2x')
plt.plot(x, 3 * x, linestyle='dashdot', label= '3x')
plt.plot(x, 4 * x, linestyle='dotted', label= '4x')

plt.title("Different Slopes")   # Insert a title
plt.xlabel("x")                 # Label the x axis
plt.ylabel("y")                 # Label the y axis
plt.legend()                    # Insert a legend
plt.show()
```

# Simple Scatter Plot 

```python
# Simple Scatter Plot-1

# Create a simple scatter plot with .plot
x = np.linspace(0, 10, 50)
y = x ** 2
plt.plot(x, y, '-o', color = 'white')
plt.show()
```

```python
# Simple Scatter Plot-2

# Create plot from scatter()
x = np.linspace(0, 10, 50)
y = rn.sample(range(1, 100), 50)
plt.scatter(x, y, marker = 'o')
plt.title("Random Plot")            # Insert a title
plt.xlabel("x")                     # Label the x axis
plt.ylabel("y")                     # Label the y axis
plt.show()
```

# Visualizing Errors

```python
# Visualizing Errors-1

# Simple error bar plot
x = np.linspace(0, 100, 60)
dy = 0.9
y = np.cos(x) + dy * np.random.randn(60)

plt.title("Cosine with Error")              # Insert a title
plt.xlabel("x")                             # Label the x axis
plt.ylabel("cos(y)")                        # Label the y axis
plt.errorbar(x, y, yerr=dy, fmt='.', ecolor='white')
plt.show()
```

# Density and Contour Plots

```python
# Density and Countour Plots-1

# Function for the z Axis
def func(x, y):
    return np.sin(x * y) * np.cos(x * y)

# Create X Y and Z
x = np.linspace(0, 10, 100)
y = np.linspace(0, 10, 75)
X, Y = np.meshgrid(x, y)
Z = func(X, Y)

#Plot the Graph
plt.title("sin(x * y) * cos(x * y)")    # Insert a title
plt.xlabel("x")                         # Label the x axis
plt.ylabel("y")                         # Label the y axis
plt.contourf(X, Y, Z, 20, cmap = 'plasma')
plt.colorbar()
plt.show()
```

# Histograms, Binnings, and Density

```python
# Histograms, Binnings, and Density-1 

# 2-D Histogram

val1 = np.random.normal(0,1, 1000)
val2 = np.random.normal(-2, 2, 1000)
val3 = np.random.normal(3, 2, 1000)

kwargs = dict(histtype='stepfilled', alpha=0.3, bins=40)

plt.hist(val1, **kwargs)
plt.hist(val2, **kwargs)
plt.hist(val3, **kwargs)
plt.show()
```

# Customizing Plot Legends

```python
# Customizing Plot Legends

# Plot four lines
plt.plot(x, x, linestyle='solid', label= 'x')
plt.plot(x, x ** 1.5, linestyle='dashed', label= 'x ^ 1.5')
plt.plot(x, x ** 2, linestyle='dashdot', label= 'x ^ 2')
plt.plot(x, x ** 2.5, linestyle='dotted', label= 'x ^ 2.5')

plt.title("Different Slopes")                       # Insert a title
plt.xlabel("x")                                     # Label the x axis
plt.ylabel("y")                                     # Label the y axis
plt.legend(framealpha = 1, frameon = True,          # Insert a legend
           fancybox = True, borderpad = 1, loc ='upper left')          
plt.show()
```

# Customizing Colorbars

```python
# Density and Countour Plots-1

# Function for the z Axis
def func(x, y):
    return np.sin(x ** y) * np.cos(y)

# Create X Y and Z
x = np.linspace(0, 10, 100)
y = np.linspace(0, 10, 75)
X, Y = np.meshgrid(x, y)
Z = func(X, Y)

#Plot the Graph
plt.title("sin(x ** y) * cos(y)")               # Insert a title
plt.xlabel("x")                                 # Label the x axis
plt.ylabel("y")                                 # Label the y axis
plt.contourf(X, Y, Z, 20, cmap = 'viridis')     # Choose this color map to translate to grey scale well.
plt.colorbar()
plt.show()
```

# Multiple Subplots

```python
# Multiple Subplots

# Create subplots
fig, ax = plt.subplots(3, 1, sharex= 'col', sharey = 'row')

# Plot on subplots
x = np.linspace(0, 100, 1000)
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
ax[2].plot(x, np.tan(x))

# Show
plt.show()
```

# Text and Annotation

```python
# Text and Annotation

#Create Plot
x = np.linspace(2, 64, 1000)
plt.plot(x, np.log2(x))
plt.title("Log2 Plot")              # Insert a title
plt.xlabel("x")                     # Label the x axis
plt.ylabel("Log2(x)")               # Label the y axis

# Annotations
plt.annotate('Log(2) = 1', xy = (2, 1),
             arrowprops=dict(facecolor='white', shrink=0.05))
plt.annotate('Log(8) = 3', xy = (8, 3),
             arrowprops=dict(facecolor='white', shrink=0.05))
plt.annotate('Log(32) = 5', xy = (32, 5),
             arrowprops=dict(facecolor='white', shrink=0.05))

# Show Plot
plt.show()
```

# Customizing Ticks

```python
# Customizing Ticks

# Plot sin(x) and 2sin(x)
fig, ax = plt.subplots()
x = np.linspace(0, 4 * np.pi, 1000)
ax.plot(x, 2 * np.sin(x), lw=3, label='2 * Sine')
ax.plot(x, np.sin(x), lw=3, label='Sine')

# Set up locators and graph
ax.xaxis.set_major_locator(plt.MultipleLocator(np.pi / 2))
ax.xaxis.set_minor_locator(plt.MultipleLocator(np.pi / 4))
ax.grid(True)
ax.legend()
ax.axis('equal')
ax.set_xlim(0, 3 * np.pi)
plt.show()
```

# Customizing Matplotlib: Configurations and Stylesheets

```python
# Customizing Matplotlib: Configurations and Stylesheets

# Same as last example with customization
def func():
    # Plot sin(x) and 2sin(x)
    fig, ax = plt.subplots()
    x = np.linspace(0, 4 * np.pi, 1000)
    ax.plot(x, 2 * np.sin(x), lw=3, label='2 * Sine')
    ax.plot(x, np.sin(x), lw=3, label='Sine')

    # Set up locators and graph
    ax.xaxis.set_major_locator(plt.MultipleLocator(np.pi / 2))
    ax.xaxis.set_minor_locator(plt.MultipleLocator(np.pi / 4))
    ax.grid(True)
    ax.legend()
    ax.axis('equal')
    ax.set_xlim(0, 3 * np.pi)
    plt.show()

#Customization
with plt.style.context('grayscale'):    
    func()
```

# Three-Dimensional Plotting in Matplotlib

```python
# Three-Dimensional Plotting in Matplotlib

def func(x, y):
    return np.cos(np.sin(x + y))

#Create graph
x = np.linspace(0, 10, 100)
y = np.linspace(0, 10, 100)
X, Y = np.meshgrid(x, y)
Z = func(X, Y)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 100, cmap='plasma')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
```

# Geographic Data with Basemap

```python
fig = plt.figure(figsize=(8, 8))

# Map view
m = Basemap(projection='lcc', resolution='l',
            width=5E6, height=5E6, 
            lat_0=33.7490, lon_0=-84.3880)

m.etopo(scale=0.5, alpha=0.5)

# Map (long, lat) to (x, y) for plotting Atlanta
x, y = m(-84.3880, 33.7490)
plt.plot(x, y, 'ok', markersize=5)
plt.text(x, y, ' Atlanta', fontsize=12)

plt.show()
```
